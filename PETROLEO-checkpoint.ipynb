```python
import numpy as np
import math 
import pandas as pd 
import seaborn as sns 
import matplotlib.pyplot as plt
import os
plt.style.use('classic')
%matplotlib inline
```

```python
basedir = '.'
df = pd.read_csv(os.path.join(basedir,'BASE DE DATOS.csv'),index_col=0, parse_dates=True)
df.head()
```

```python
df.plot()
```

```python
fig = plt.figure(figsize=(12,8), tight_layout=True)


ax1 = fig.add_subplot(3,2,1)
ax1.plot(df.PPMM,'-',lw=3,alpha = 0.7)
ax1.grid()
ax1.set_title('Precio de Petroleo: mezcla mexicana')
ax1.set_ylabel('Dolares por '+' barril')


ax2 = fig.add_subplot(3,2,2)
ax2.plot(df.XPP,'-',lw=3,alpha = 0.7)
ax2.grid()
ax2.set_title('Exportaciones de Petroleo')
ax2.set_ylabel('Miles de '+' dolares')

ax3 = fig.add_subplot(3,2,3)
ax3.plot(df.DP,'-',lw=3,alpha = 0.7)
ax3.grid()
ax3.set_title('Petroleo Crudo')
ax3.set_ylabel('Miles de '+' dolares')

ax4 = fig.add_subplot(3,2,4)
ax4.plot(df.GN,'-',lw=3,alpha = 0.7)
ax4.grid()
ax4.set_title('Gas Natural')
ax4.set_ylabel('Miles de '+' dolares')

ax5 = fig.add_subplot(3,2,5)
ax5.plot(df.DP,'-',lw=3,alpha = 0.7)
ax5.grid()
ax5.set_title('Derivados de Petroleo')
ax5.set_ylabel('Miles de '+' dolares')

ax6 = fig.add_subplot(3,2,6)
ax6.plot(df.MPP,'-',lw=3,alpha = 0.7)
ax6.grid()
ax6.set_title('Importaciones de Petroleo')
ax6.set_ylabel('Miles de '+' dolares')


fig.tight_layout()
```

```python
%matplotlib inline

import statsmodels.api as sm
```

```python
fig, ax = plt.subplots(figsize = (10,6))
df['PPMM'].plot(ax=ax)
legend = ax.legend(loc='best')
plt.grid()
```

```python
PPMM_cycle, PPMM_trend = sm.tsa.filters.hpfilter(df.PPMM, lamb=1600)
```

```python
PPMM_decomp = df[['PPMM']].copy()
PPMM_decomp["cycle"] = PPMM_cycle
PPMM_decomp["trend"] = PPMM_trend
```

```python
fig = plt.figure(figsize=(10,6))
ax = fig.add_subplot(111)
PPMM_decomp[["PPMM", "trend"]]["feb-03":].plot(ax=ax, fontsize=16);
legend = ax.get_legend()
plt.grid()
```
